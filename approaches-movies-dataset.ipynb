```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
df = pd.read_csv("/kaggle/input/tmdb-top-rated-movies-dataset/movie_dataset.csv")
df.head()
df.columns
```

```python
df["original_language"].unique()
df["original_language"].value_counts().head(10)
```

In order to include it in the transactions more easily, I prefer to separate the date by year below.

```python
df["release_date"] = df["release_date"].astype(str).str.split("-")
df.head()
```

```python
df["release_date"] = df["release_date"].apply(lambda x: x[0])

df.head()
```

```python
df = df.sort_values(by="vote_count", ascending=False)
df.reset_index(drop=True).head()
```

```python
df.groupby("original_language")[["vote_count"]].count().sort_values(by="vote_count", ascending=False).head()
```

```python
df.groupby("genres")[["popularity"]].count().sort_values(by="popularity", ascending=False).head(10)
```
