```python
bins_=[0, 7, 30, 90, df["SaleCheckInDayDiff"].max()]
labels_=["Last Minuters", "Potential Planners", "Planners", "Early Bookers"]
```

```python
df["CheckInType"] = pd.cut(df["SaleCheckInDayDiff"], bins=bins_, labels=labels_, include_lowest=True)
```

```python
df["CheckIn2"] = pd.qcut(df["SaleCheckInDayDiff"], q=4, labels=labels_)
```

```python
df.head()
```

```python
df.drop("CheckIn2", axis=1, inplace=True)
df.head()
```

```python
df.groupby(["SaleCityName", "ConceptName", "CheckInType"], observed=True).agg({"Price": ["mean", "count"]})
```

```python
agg_df = df.groupby(["SaleCityName", "ConceptName", "Seasons"]).agg({"Price" : "mean"}).sort_values(by="Price", ascending=False)
```

```python
agg_df.reset_index(inplace=True)
agg_df.head()
```

```python
df["sales_level_based"] = df[["SaleCityName", "ConceptName", "Seasons"]].apply(lambda x: "_".join(x), axis=1)
df["sales_level_based"] = df["sales_level_based"].str.upper()
df["sales_level_based"].head()
```

```python
agg_df = df[["sales_level_based", "Price"]]
agg_df.head()
```
