```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import seaborn as sns
df = sns.load_dataset("titanic")
```

```python
##Aggregation & Grouping
df.groupby("sex")["age"].mean()
```

```python
df.groupby(["sex", "embark_town"]).agg({"age": ["mean", "sum"],
                      "survived": "mean"})
```

```python
##Pivot Table; df.pivot_table("kesişim", "satır", "sütun")
df.pivot_table("survived", "sex", "embarked")
```

**** Pivot table ın ön tanımlı değeri mean(ortalama)dır. Biçimlendirmek için sona ör; aggfunc="std" eklenir.

**** Sayısal değişkeni kategorik değişkene çevirmek; Gurpları tanımlayabiliyorsan pd.cut tanımlayamıyorsan pd.qcut kullanılır.


```python
df["new_age"] = pd.cut(df["age"], [0, 10, 18, 25, 40, 90])
df.pivot_table("survived", "sex", ["new_age", "class"])
```
